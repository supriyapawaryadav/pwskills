# 1.What is hypothesis testing in statistics

Hypothesis testing in statistics is a method used to make inferences or draw conclusions about a population based on a sample of data. It involves formulating two competing statements: the null hypothesis (H0) and the alternative hypothesis (H1 or H).

# 2. What is the null hypothesis, and how does it differ from the alternative hypothesis

Null Hypothesis (H0): This is a statement that there is no effect or no difference in the population. It represents the default assumption or status quo.

Alternative Hypothesis (H1or Ha): This is a statement that contradicts the null hypothesis, indicating that there is an effect or difference.

# 3.What is the significance level in hypothesis testing, and why is it important


The significance level plays a critical role in determining whether the data provide strong enough evidence to reject the null hypothesis and supports the decision-making process in statistical tests.

1.Control over Type I Error

2.Determines Evidence Required

3.Interpretation of Results


0.05: The most commonly used significance level, meaning you accept a 5% chance of making a Type I error.

0.01: A more stringent threshold, reducing the chance of Type I errors to 1%.

0.10: A less stringent threshold, allowing for a 10% chance of a Type I error, often used in exploratory research.

# 4.What does a P-value represent in hypothesis testing?

p-value (probability value) represents the probability of observing the data, or something more extreme, given that the null hypothesis is true. Essentially, it tells you how compatible your data is with the null hypothesis.

# 5.How do you interpret the P-value in hypothesis testing?

P-value < α (typically 0.05):

Example: If you are testing whether a new drug is effective and your p-value is 0.02 with a significance level of 0.05, you reject the null hypothesis, concluding that the drug has a statistically significant effect.

P-value ≥ α (typically 0.05):

Example: If you test the same drug and get a p-value of 0.08 with a significance level of 0.05, you fail to reject the null hypothesis, suggesting that the data does not provide enough evidence to conclude the drug is effective.

# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

```python
from scipy.stats import norm

# Inputs
sample_mean = 105
population_mean = 100
population_std = 15
sample_size = 30
significance_level = 0.05

# Z-test calculations
standard_error = population_std / (sample_size ** 0.5)
z_score = (sample_mean - population_mean) / standard_error
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Results and interpretation
print(f"Z-Score: {z_score:.3f}")
print(f"P-Value: {p_value:.3f}")

if p_value < significance_level:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

# 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean

```python
from scipy.stats import norm
import numpy as np

# Sample data
sample = [48, 50, 52, 47, 49, 51, 46, 53, 50, 49]  # Example sample

# Known population parameters
population_mean = 50
population_std = 5  # Population standard deviation (assumed known)

# Calculate sample statistics
sample_mean = np.mean(sample)
sample_size = len(sample)

# Calculate the Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Known parameters
population_mean = 50
population_std = 5
sample = [48, 50, 52, 47, 49, 51, 46, 53, 50, 49]

# Sample statistics
sample_mean = np.mean(sample)
sample_size = len(sample)

# Calculate Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate critical Z-value for a two-tailed test (alpha = 0.05)
alpha = 0.05
z_critical = norm.ppf(1 - alpha / 2)

# Visualization
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Standard Normal Distribution", color="blue")

# Shade critical regions
plt.fill_between(x, y, 0, where=(x <= -z_critical) | (x >= z_critical), color="red", alpha=0.3, label="Rejection Region")

# Plot Z-statistic
plt.axvline(z_stat, color="green", linestyle="--", label=f"Z-statistic = {z_stat:.2f}")
plt.axvline(-z_critical, color="red", linestyle="--", label=f"Critical Z = ±{z_critical:.2f}")
plt.axvline(z_critical, color="red", linestyle="--")

# Add labels and legend
plt.title("Two-tailed Z-test: Decision Regions")
plt.xlabel("Z-value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()

# Show plot
plt.show()

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.2f}")
if abs(z_stat) > z_critical:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(mu_null, mu_alternative, std, sample_size, alpha=0.05):
    """
    Visualizes Type 1 and Type 2 errors for hypothesis testing.

    Parameters:
    - mu_null: Mean under the null hypothesis.
    - mu_alternative: Mean under the alternative hypothesis.
    - std: Population standard deviation.
    - sample_size: Number of observations in the sample.
    - alpha: Significance level (default = 0.05).
    """
    # Standard error
    se = std / np.sqrt(sample_size)
    
    # Critical values
    z_critical = norm.ppf(1 - alpha / 2)  # Two-tailed test
    x_critical_low = mu_null - z_critical * se
    x_critical_high = mu_null + z_critical * se

    # X values for plotting
    x = np.linspace(mu_null - 4*se, mu_alternative + 4*se, 1000)

    # Null and alternative distributions
    null_dist = norm.pdf(x, mu_null, se)
    alt_dist = norm.pdf(x, mu_alternative, se)

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, null_dist, label="Null Hypothesis (H0)", color="blue")
    plt.plot(x, alt_dist, label="Alternative Hypothesis (H1)", color="orange")
    
    # Shade Type 1 Error (alpha)
    plt.fill_between(x, null_dist, 0, where=(x <= x_critical_low) | (x >= x_critical_high), 
                     color="red", alpha=0.3, label="Type 1 Error (α)")
    
    # Shade Type 2 Error (beta)
    plt.fill_between(x, alt_dist, 0, where=(x > x_critical_low) & (x < x_critical_high), 
                     color="green", alpha=0.3, label="Type 2 Error (β)")
    
    # Add vertical lines for critical values
    plt.axvline(x_critical_low, color="red", linestyle="--", label=f"Critical Value Low = {x_critical_low:.2f}")
    plt.axvline(x_critical_high, color="red", linestyle="--", label=f"Critical Value High = {x_critical_high:.2f}")

    # Labels and legend
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

    # Print critical values
    print(f"Critical values for rejection: {x_critical_low:.2f}, {x_critical_high:.2f}")

# Example Usage
mu_null = 50          # Null hypothesis mean
mu_alternative = 52   # Alternative hypothesis mean
std = 10              # Population standard deviation
sample_size = 30      # Sample size

visualize_type1_type2(mu_null, mu_alternative, std, sample_size)
```

# 9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
   
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for sample)
    n = len(data)  # Sample size
    t_critical = t.ppf(1 - (1 - confidence) / 2, df=n - 1)  # t-critical value for the confidence level
    
    # Calculate the margin of error
    margin_of_error = t_critical * (sample_std / np.sqrt(n))
    
    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return lower_bound, upper_bound

# Example Usage
sample_data = [45, 50, 55, 60, 65, 70]
lower, upper = confidence_interval(sample_data)
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
# Bayesian Inference using Bayes' Theorem

# Prior probabilities
P_biased = 0.5  # Prior belief that the coin is biased
P_fair = 0.5    # Prior belief that the coin is fair

# Likelihoods
P_heads_given_biased = 0.7  # Likelihood of heads given biased coin
P_heads_given_fair = 0.5    # Likelihood of heads given fair coin

# Evidence: Probability of observing 7 heads out of 10 flips (total probability of heads data)
P_heads_data = (P_heads_given_biased * P_biased) + (P_heads_given_fair * P_fair)

# Bayes' Theorem to compute the posterior probabilities
P_biased_given_data = (P_heads_given_biased * P_biased) / P_heads_data
P_fair_given_data = (P_heads_given_fair * P_fair) / P_heads_data

# Display the posterior probabilities
print(f"Posterior probability that the coin is biased: {P_biased_given_data:.4f}")
print(f"Posterior probability that the coin is fair: {P_fair_given_data:.4f}")
```

# 13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

```python
import numpy as np

def calculate_expected_frequencies(observed_data):
    
    # Convert observed data into a numpy array
    observed_data = np.array(observed_data)
    
    # Calculate row totals, column totals, and grand total
    row_totals = observed_data.sum(axis=1, keepdims=True)
    column_totals = observed_data.sum(axis=0, keepdims=True)
    grand_total = observed_data.sum()
    
    # Calculate the expected frequencies using the formula
    expected_frequencies = (row_totals * column_totals) / grand_total
    
    return expected_frequencies

# Example observed data (Contingency Table)
observed_data = [
    [30, 10],  # Male: 30 Coffee, 10 Tea
    [20, 25],  # Female: 20 Coffee, 25 Tea
]

# Calculate expected frequencies
expected = calculate_expected_frequencies(observed_data)

# Display the observed and expected frequencies
print("Observed Frequencies:")
print(np.array(observed_data))
print("\nExpected Frequencies:")
print(expected)
```

# 15.Create a Python script to simulate and visualize the Chi-square distribution and discuss its characteristics

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Function to simulate and visualize Chi-square distribution
def plot_chi_square_distribution(degrees_of_freedom, num_samples=10000):
    # Simulate Chi-square data
    chi_square_data = chi2.rvs(df=degrees_of_freedom, size=num_samples)
    
    # Plot the histogram of the simulated data
    plt.figure(figsize=(8, 6))
    plt.hist(chi_square_data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')
    
    # Plot the theoretical Chi-square distribution curve
    x = np.linspace(0, np.max(chi_square_data), 100)
    plt.plot(x, chi2.pdf(x, df=degrees_of_freedom), 'r-', lw=2, label=f'Chi-square df={degrees_of_freedom}')
    
    # Add labels and title
    plt.title(f'Chi-square Distribution (df={degrees_of_freedom})', fontsize=14)
    plt.xlabel('Value', fontsize=12)
    plt.ylabel('Density', fontsize=12)
    plt.legend(loc='upper right')
    
    # Display the plot
    plt.show()

# Simulate and visualize Chi-square distributions for different degrees of freedom
for df in [1, 2, 5, 10]:
    plot_chi_square_distribution(df)
```

# 16. Implement an F-test using Python to compare the variances of two random samples

```python
import numpy as np
from scipy.stats import f

# Function to perform F-test for comparing variances of two samples
def f_test(sample1, sample2):
    # Calculate sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    
    # F-statistic is the ratio of the two variances
    f_statistic = var1 / var2 if var1 > var2 else var2 / var1
    
    # Determine the degrees of freedom
    df1 = len(sample1) - 1  # Degrees of freedom for sample 1
    df2 = len(sample2) - 1  # Degrees of freedom for sample 2
    
    # Calculate the p-value using the F-distribution
    p_value = 1 - f.cdf(f_statistic, df1, df2)
    
    # Print the results
    print(f"F-statistic: {f_statistic:.4f}")
    print(f"p-value: {p_value:.4f}")
    
    # Interpretation based on p-value
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The variances are not significantly different.")

# Example data: Two samples with different variances
sample1 = np.random.normal(loc=50, scale=10, size=100)  # Sample 1 with standard deviation 10
sample2 = np.random.normal(loc=50, scale=20, size=100)  # Sample 2 with standard deviation 20

# Perform F-test
f_test(sample1, sample2)
```

# 19.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

Normality Check:

Shapiro-Wilk test is used to test whether the data in each group follows a normal distribution.
If the p-value is less than 0.05, we reject the null hypothesis (H₀) and conclude that the data is not normally distributed.

Equal Variance Check:

Levene's test is used to check if the variances of the groups are equal.
If the p-value is less than 0.05, we reject the null hypothesis (H₀) and conclude that the variances are unequal.

Independence Check:

Durbin-Watson statistic is used to check for autocorrelation in residuals. The statistic should ideally be between 1.5 and 2.5 for independent data.
Values outside of this range suggest that the data may have autocorrelation (i.e., non-independence).

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Function to check assumptions for ANOVA
def check_anova_assumptions(*groups):
    # 1. Normality Check using Shapiro-Wilk Test
    print("\nNormality Check (Shapiro-Wilk Test):")
    for i, group in enumerate(groups, 1):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i}: W-statistic = {stat:.4f}, p-value = {p_value:.4f}")
        if p_value < 0.05:
            print(f"Group {i}: Data is NOT normally distributed (reject H0).")
        else:
            print(f"Group {i}: Data is normally distributed (fail to reject H0).")
    
    # 2. Equal Variance Check using Levene's Test
    print("\nEqual Variance Check (Levene's Test):")
    stat, p_value = stats.levene(*groups)
    print(f"Levene’s test statistic = {stat:.4f}, p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("The variances are significantly different (reject H0).")
    else:
        print("The variances are equal (fail to reject H0).")

    # 3. Independence Check using plotting (Visual check for outliers and independence)
    print("\nIndependence Check (Visual inspection using plots):")
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=[group for group in groups])
    plt.title("Boxplot for Independence Check (Visual Inspection for Outliers)")
    plt.xlabel("Groups")
    plt.ylabel("Values")
    plt.show()

    print("\nIndependence is assumed if there is no clear pattern in the boxplot and no outliers in any group.")

# Example data for three groups
group1 = np.random.normal(loc=50, scale=10, size=30)  # Group 1
group2 = np.random.normal(loc=55, scale=10, size=30)  # Group 2
group3 = np.random.normal(loc=60, scale=10, size=30)  # Group 3

# Check assumptions for ANOVA
check_anova_assumptions(group1, group2, group3)
```

# 20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt

# Generate some example data
np.random.seed(42)

# Factor 1: Diet (3 levels: A, B, C)
diet = np.repeat(['A', 'B', 'C'], 20)

# Factor 2: Exercise (2 levels: Low, High)
exercise = np.tile(['Low', 'High'], 30)

# Dependent variable: Weight Loss (in kg)
weight_loss = np.random.normal(5, 1.5, 60) + (diet == 'A')*2 + (exercise == 'High')*3 + (diet == 'B')*1 + (exercise == 'Low')*0.5

# Create a DataFrame
data = pd.DataFrame({'Diet': diet, 'Exercise': exercise, 'WeightLoss': weight_loss})

# Perform Two-Way ANOVA
model = ols('WeightLoss ~ C(Diet) * C(Exercise)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Display ANOVA table
print(anova_table)

# Visualize the interaction effects using a line plot
plt.figure(figsize=(8, 6))
sns.lineplot(x='Diet', y='WeightLoss', hue='Exercise', data=data, markers=['o', 's'], style='Exercise')
plt.title("Interaction between Diet and Exercise on Weight Loss")
plt.xlabel("Diet Type")
plt.ylabel("Weight Loss (kg)")
plt.legend(title="Exercise Level")
plt.show()
```

# 21.Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing

The F-distribution is crucial in ANOVA and variance comparison tests. This visualization helps understand its shape, and its role in hypothesis testing, especially when comparing the variances of two or more groups.

Hypothesis Testing:

The F-statistic is used to test the null hypothesis in various statistical tests (e.g., ANOVA). In ANOVA, the null hypothesis typically states that all group means are equal.
If the calculated F-statistic is greater than a critical value from the F-distribution for a given significance level (e.g., α = 0.05), we reject the null hypothesis, suggesting that there are significant differences between the groups.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters for F-distribution
dfn = 5  # Degrees of freedom for the numerator (between-group variance)
dfd = 2  # Degrees of freedom for the denominator (within-group variance)

# Generate x values from 0 to 5
x = np.linspace(0, 5, 1000)

# Calculate the corresponding F-distribution values for the given degrees of freedom
f_pdf = stats.f.pdf(x, dfn, dfd)

# Plotting the F-distribution
plt.figure(figsize=(8, 6))
plt.plot(x, f_pdf, label=f'F-distribution (dfn={dfn}, dfd={dfd})', color='blue')
plt.title('F-distribution with dfn=5 and dfd=2')
plt.xlabel('F-statistic')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend(loc='upper right')
plt.show()

# Additional explanation
print(f"The F-distribution with numerator df={dfn} and denominator df={dfd} is used to test the hypothesis of equal variances.")
print("The F-statistic is the ratio of two variances: variance between group means (numerator) / variance within groups (denominator).")
print("If the F-statistic is large, it suggests that the between-group variance is larger than the within-group variance, indicating a significant difference.")
```

# 22.Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Example data for three groups
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)  # Group 1
group2 = np.random.normal(loc=55, scale=10, size=30)  # Group 2
group3 = np.random.normal(loc=60, scale=10, size=30)  # Group 3

# Combine data into a DataFrame for easy visualization and ANOVA
data = pd.DataFrame({
    'Value': np.concatenate([group1, group2, group3]),
    'Group': ['Group 1'] * len(group1) + ['Group 2'] * len(group2) + ['Group 3'] * len(group3)
})

# Perform One-Way ANOVA
f_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Display the ANOVA result
print(f"ANOVA result: F-statistic = {f_statistic:.4f}, p-value = {p_value:.4f}")

# Interpret the p-value
if p_value < 0.05:
    print("There is a significant difference between the group means (reject H0).")
else:
    print("There is no significant difference between the group means (fail to reject H0).")

# Visualization with Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='Value', data=data, palette="Set3")
plt.title("One-Way ANOVA: Comparison of Group Means")
plt.xlabel("Group")
plt.ylabel("Value")
plt.show()
```

# 24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import numpy as np
import scipy.stats as stats

# Step 1: Simulate data for a sample
np.random.seed(42)
sample_size = 30
population_variance = 25  # Hypothesized population variance (σ₀²)
sample_data = np.random.normal(loc=50, scale=np.sqrt(population_variance), size=sample_size)

# Step 2: Calculate the sample variance
sample_variance = np.var(sample_data, ddof=1)

# Step 3: Perform Chi-Square Test for Population Variance
# Hypothesized variance
sigma_0_squared = 25  # Hypothesized population variance (σ₀²)

# Chi-square statistic formula: (n-1) * s² / σ₀²
chi_square_statistic = (sample_size - 1) * sample_variance / sigma_0_squared

# Step 4: Calculate the p-value (two-tailed test)
df = sample_size - 1  # Degrees of freedom
p_value = 2 * min(stats.chi2.cdf(chi_square_statistic, df), 1 - stats.chi2.cdf(chi_square_statistic, df))

# Step 5: Display the results
alpha = 0.05  # Significance level
print(f"Chi-Square Statistic: {chi_square_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: The sample variance is significantly different from the hypothesized population variance.")
else:
    print("Fail to reject the null hypothesis: The sample variance is not significantly different from the hypothesized population variance.")
```

# 25.Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

```python
import numpy as np
import scipy.stats as stats

# Step 1: Simulate two datasets for comparison (successes and sample sizes)
# Group 1
successes_1 = 50
sample_size_1 = 100

# Group 2
successes_2 = 60
sample_size_2 = 120

# Step 2: Calculate the sample proportions
p1 = successes_1 / sample_size_1
p2 = successes_2 / sample_size_2

# Step 3: Calculate the pooled proportion
pooled_proportion = (successes_1 + successes_2) / (sample_size_1 + sample_size_2)

# Step 4: Calculate the Z-statistic
z_statistic = (p1 - p2) / np.sqrt(pooled_proportion * (1 - pooled_proportion) * (1/sample_size_1 + 1/sample_size_2))

# Step 5: Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Step 6: Display the results
alpha = 0.05  # Significance level
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < alpha:
    print("Reject the null hypothesis: The proportions of the two groups are significantly different.")
else:
    print("Fail to reject the null hypothesis: The proportions of the two groups are not significantly different.")
```

# 26.Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate two datasets with different variances
np.random.seed(42)

# Dataset 1 (larger variance)
data1 = np.random.normal(loc=50, scale=10, size=100)

# Dataset 2 (smaller variance)
data2 = np.random.normal(loc=50, scale=5, size=100)

# Step 2: Calculate the sample variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Step 3: Calculate the F-statistic
F_statistic = var1 / var2  # Assuming var1 > var2

# Step 4: Calculate the degrees of freedom for both samples
df1 = len(data1) - 1  # Degrees of freedom for sample 1
df2 = len(data2) - 1  # Degrees of freedom for sample 2

# Step 5: Calculate the p-value using the F-distribution
p_value = 1 - stats.f.cdf(F_statistic, df1, df2)  # Right-tail test

# Step 6: Display the results
alpha = 0.05  # Significance level
print(f"F-statistic: {F_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the results
if p_value < alpha:
    print("Reject the null hypothesis: The variances of the two datasets are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances of the two datasets are not significantly different.")

# Visualization: Histograms of the two datasets
plt.figure(figsize=(10, 6))
plt.hist(data1, bins=20, alpha=0.7, label='Dataset 1 (larger variance)', color='skyblue', edgecolor='black')
plt.hist(data2, bins=20, alpha=0.7, label='Dataset 2 (smaller variance)', color='salmon', edgecolor='black')
plt.title('Histograms of Two Datasets')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Visualization of the F-distribution for critical value
F_critical = stats.f.ppf(1 - alpha, df1, df2)

plt.figure(figsize=(10, 6))
x = np.linspace(0, 5, 500)
y = stats.f.pdf(x, df1, df2)
plt.plot(x, y, label='F-distribution', color='green')
plt.axvline(F_critical, color='red', linestyle='--', label=f'Critical Value (F = {F_critical:.4f})')
plt.title('F-distribution and Critical Value')
plt.xlabel('F-statistic')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
